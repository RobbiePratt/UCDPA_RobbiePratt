### 3.1.  Your project should use Regex to extract a pattern in data - It was decided to extract the  and date seperately from the CreationDate Column using Regex


#### Import Python Packages

```python
import pandas as pd
import re
```

#### Import a CSV file into a Pandas DataFrame

```python
df_DateRegEx=pd.read_csv('valid.csv')
df_DateRegEx
```

Make extra column for time

```python
df_DateRegEx ['Time'] = 0
df_DateRegEx ['Date'] = 0
```

```python
df_DateRegEx
```

Set Indexes

```python
index_creationdate = df_DateRegEx.columns.get_loc('CreationDate')
index_time = df_DateRegEx.columns.get_loc('Time')
index_date = df_DateRegEx.columns.get_loc('Date')

print(index_creationdate, index_time, index_date)
```

Define Pattern to be extracted

```python
time_pattern = r'([0-9]{2}\:[0-9]{2}\:[0-9]{2})'
date_pattern = r'([0-9]{4}\-[0-9]{2}\-[0-9]{2})'
```

Look for data in dateframe ( Make use of iterators)

```python
for row in range (0, len (df_DateRegEx)):
    time = re.search(time_pattern, df_DateRegEx.iat[row, index_creationdate]).group()
    df_DateRegEx.iat[row, index_time] = time
    date = re.search(date_pattern, df_DateRegEx.iat[row, index_creationdate]).group()
    df_DateRegEx.iat[row, index_date] = date
```
